# Implementación del método Monte Carlo 

Aproximar la integral:

$$
\int_0^{\pi}sin(x)dx
$$

Límites de integración


```python
a=0
b=np.pi
N=1000
```

```python
b
```

```python
A=np.zeros(N)
```

```python
A
```

```python
np.random.uniform(a,b)
```

```python
for i in range(1000):
    A[i]=np.random.uniform(a,b)
```

```python
A
```

```python
np.cos(1)
```

```python
integral=0.
def f(x):
    return np.sin(x)
for i in A:
    integral=integral+f(i)
```

```python
I=((b-a)/N)*integral
```
