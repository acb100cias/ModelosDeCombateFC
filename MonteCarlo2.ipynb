```python
import numpy as np
import matplotlib.pyplot as plt
import sympy
```

```python
a=0
b=np.pi
N=1000
A=np.zeros(N)
for i in range(N):
    A[i]=np.random.uniform(a,b)
suma=0.
for j in A:
    suma+=np.sin(j)
integral=((b-a)/N)*suma
    
```

```python
L=[]
N=1000
for k in range(1000):
    A=np.zeros(N)
    for i in range(N):
        A[i]=np.random.uniform(a,b)
    suma=0.
    for j in A:
        suma+=np.sin(j)
    integral=((b-a)/N)*suma
    L.append(integral)

```

```python
plt.hist(L,bins=30,ec='k')
plt.show()
```

```python
a=1
b=5
L2=[]
for k in range(1000):
    A=np.zeros(N)
    def f(x):
        return x**2
    for i in range(N):
        A[i]=np.random.uniform(a,b)
    suma=0.
    for j in A:
        suma+=f(j)
    integral=((b-a)/N)*suma
    L2.append(integral)
plt.hist(L2,bins=30,ec='k')
plt.show()
```
